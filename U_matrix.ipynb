```python
import numpy as np
```

```python
def generate_U_Uck(N_tot, N, m, n):
    for k in range(N_tot):    
        p = k%n     
        Uck = np.zeros((N, 1))
        Uck[:n-p] = 1
        Uk = np.zeros((N, m-1))
        for i in range((N+p)//n):
            Uk[(i+1)*n-p:(i+2)*n-p, i] = 1
        
        yield Uk, Uck
```

```python
N_tot = 20
N = 10
m = 5
n = 3
generator = generate_U_Uck(N_tot, N, m, n)

for k in range(10):
    print(k)
    Uk, Uck = next(generator)
    print(np.hstack((Uck, Uk)))
```
